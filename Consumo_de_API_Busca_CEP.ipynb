```python
# Importando Bibliotecas
import requests
import pandas as pd
```

```python
# Função para fazer a requisição para um CEP específico
def busca_cep(cep):
    url = f'https://viacep.com.br/ws/{cep}/json/'
    resposta = requests.get(url)
    if resposta.status_code == 200:
        json_data = resposta.json()
        return json_data
    else:
        print(f'Falha na requisição para o CEP {cep}. Código de status:', resposta.status_code)
        return None

# Lista para armazenar os dados dos CEPs
resposta_ceps = []

# Loop para solicitar o CEP e adicionar os dados ao DataFrame
while True:
    cep_digitado = input('Digite um CEP (ou digite "sair" para encerrar): ')
    if cep_digitado.lower() == 'sair':
        break
    resposta_cep = busca_cep(cep_digitado)
    if resposta_cep:
        resposta_ceps.append(resposta_cep)

# Criando o DataFrame a partir dos dados dos CEPs
df_cep = pd.DataFrame(resposta_ceps)
df_cep = df_cep.drop(['gia', 'siafi', 'ibge', 'ddd'], axis=1)
```
